```python
car_raw.dropna(axis=0,inplace=True)`
```

## probability ditribution Exploration

```python
car_raw.describe(include='all')
```

```python
plt.figure(figsize=(8,5))
sns.distplot(car_raw['Price'])
plt.title('Car Price Distribution')
```

```python
q=car_raw['Price'].quantile(q=0.99)
q
```

```python
car_1=car_raw[car_raw['Price']<q]
car_1.describe()
```

```python
plt.figure(figsize=(14,6))

plt.subplot(1,2,1)
sns.distplot(car_1['Price']);
plt.title('After Removing Outliars');

plt.subplot(1,2,2)
sns.distplot(car_raw['Price']);
plt.title('With Outliars');
```

```python
plt.figure(figsize=(18,7))

plt.subplot(1,3,1)
sns.distplot(car_1['Mileage']);

plt.subplot(1,3,2)
sns.distplot(car_1['EngineV'])

plt.subplot(1,3,3)
sns.distplot(car_1['Year']);
```

```python
q=car_1['Mileage'].quantile(0.99)
car_2=car_1[car_1['Mileage']<q]
```

```python
sns.distplot(car_2['Mileage']);
plt.title('After Removing Outliars (Mileage)');
```

```python
car_3=car_2[car_2['EngineV']<6.5]
```

```python
sns.distplot(car_3['EngineV']);
plt.title('After Removing Outliars (Car Engine Volume)');
```

```python
q = car_3['Year'].quantile(0.01);
car_4 = car_3[car_3['Year']>q]

sns.distplot(car_4['Year']);
plt.title('Year after removing outliers');
```
